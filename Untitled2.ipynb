```python
import random

# Define the Player and Enemy classes
class Character:
    def __init__(self, name, hp, attack, defense):
        self.name = name
        self.hp = hp
        self.attack = attack
        self.defense = defense
        self.level = 1
        self.xp = 0

    def takeDamage(self, damage):
        damageTaken = max(0, damage - self.defense)
        self.hp -= damageTaken
        print(f"{self.name} takes {damageTaken} damage! Remaining HP: {self.hp}")
        return self.hp

    def gainXp(self, xp):
        self.xp += xp
        if self.xp >= 10 * self.level:
            self.levelUp()

    def levelUp(self):
        self.level += 1
        self.hp += 10
        self.attack += 2
        self.defense += 2
        print(f"{self.name} leveled up! Now at level {self.level}")

class Player(Character):
    def __init__(self, name):
        super().__init__(name, 100, 10, 5)

class Enemy(Character):
    def __init__(self, name, level):
        super().__init__(name, 20 * level, 5 * level, 2 * level)

# Define the rock-paper-scissors mechanic
def rps(playerChoice):
    choices = ["rock", "paper", "scissors"]
    enemyChoice = random.choice(choices)
    print(f"Enemy chose {enemyChoice}")

    if playerChoice == enemyChoice:
        return 0  # Draw
    elif (playerChoice == "rock" and enemyChoice == "scissors") or \
         (playerChoice == "paper" and enemyChoice == "rock") or \
         (playerChoice == "scissors" and enemyChoice == "paper"):
        return 1  # Player wins
    else:
        return -1  # Enemy wins

# Main game loop
def gameLoop():
    playerName = input("Enter your character's name: ")
    player = Player(playerName)

    print(f"Welcome, {player.name}!.")

    level = 1
    while True:
        enemy = Enemy(f"Enemy Level {level}", level)
        print(f"A wild {enemy.name} appears!")

        while enemy.hp > 0 and player.hp > 0:
            print(f"\n{player.name} HP: {player.hp}")
            print(f"{enemy.name} HP: {enemy.hp}")

            playerChoice = input("Choose rock, paper, or scissors: ").lower()
            if playerChoice not in ["rock", "paper", "scissors"]:
                print("Invalid choice, try again.")
                continue

            result = rps(playerChoice)
            if result == 1:
                print("You won the round!")
                damage = player.attack
                enemy.takeDamage(damage)
            elif result == -1:
                print("You lost the round!")
                damage = enemy.attack
                playerAlive = player.takeDamage(damage)
                if not playerAlive:
                    print("You were defeated! Game over.")
                    return
            else:
                print("It's a draw!")

        if enemy.hp <= 0:
            print(f"You defeated {enemy.name}!")
            xp = level * 5
            player.gainXp(xp)
            print(f"You gained {xp} XP.")
            level += 1

        print("\n---\n")

gameLoop()
```
